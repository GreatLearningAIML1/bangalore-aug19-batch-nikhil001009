### Q2.Prepare the model data by converting non-numeric to dummy ( 1 Marks)
##### Hint: Use get_dummies

```python
credit_df.info()
```

```python
# Print Shape of model data
```

```python
categorical_columns = []
for column in list(credit_df.columns):
    if credit_df[column].dtype == 'O':
        categorical_columns.append(column)
```

```python
credit_df_dummies = pd.get_dummies(credit_df[categorical_columns])
credit_df_dummies.head()
```

### Check for highly correlated variables but don't required any treatment for this use case

```python
credit_df.corr().loc['default']
```

### Drop the original variables which are converted to dummy

```python
credit_df = credit_df.select_dtypes([np.number])
credit_df.head()
credit_df = pd.concat([credit_df, credit_df_dummies], axis=1)
credit_df.head()
```

### Q3 Split Train/Test data 70:30 ratio( 1 Marks)
##### Hint:from sklearn.model_selection import train_test_split

```python
from sklearn.model_selection import train_test_split

X =  credit_df.drop("default", axis=1)
y =  credit_df.pop("default")

train_set, test_set, train_labels, test_labels = train_test_split(X, y, test_size=0.30, random_state=1)
```

### Q4 Build Random Forest Model( 1 Marks)
#### Hint:from sklearn.ensemble import RandomForestClassifier using n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_jobs=2,n_estimators=500,criterion="entropy",random_state=9999)
rfcl = rfcl.fit(train_set, train_labels)
test_pred = rfcl.predict(test_set)
rfcl.score(test_set , test_labels)
```

### Q5 Calculate Confusion Matrix and Accuracy score (1 Marks)
##### Hint: Use confusion_matrix and accuracy_score

```python
from sklearn import metrics
print(metrics.confusion_matrix(test_labels, test_pred))
rfcl.score(test_set , test_labels)
```

### Q6 Show the list of the features importance( 1 Marks)

```python
print(pd.DataFrame(rfcl.feature_importances_, columns=["Importance"], index = test_set.columns))
```

### Q7 K-fold cross-validation( 2 Marks)
##### k-fold cross validation( without stratification)
##### Usually k is set as 10-20 in practical settings, depends on data set size

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Use below values
num_folds = 10
seed = 77
```

```python
#Validate the Random Forest model build above using k fold
```

```python
kfold = KFold(n_splits=num_folds, random_state=seed)
results = cross_val_score(rfcl, X, y, cv=kfold)
print(results)
```

```python
#Calculate Mean score
```

```python
print("The mean score: " + str(results.mean()))
```

```python
# Calculate score standard deviation using std()
```

```python
print("The Standard deviation is: " + str(results.std()))
```

# Q8 Print the confusion matrix( 1 Marks)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score, cross_val_predict
y_predict_RF = cross_val_predict(rfcl, X, y, cv=kfold)
print("The confusion matrix of Random Forest from the K-fold cross validation:")
print(metrics.confusion_matrix(y, y_predict_RF))
```

# Q9.Classification accuracy: 
percentage of correct predictions and Calculate sensitivity (or True Positive Rate or Recall) and Precision.

```python
print("The classification accuracy report for Random Forest after K-fold Cross Validation:")
print(metrics.classification_report(y, y_predict_RF))
```
